```python
from random import randint

word_list = {
    'word_1': 'sadik',
    'word_2': 'mango',
    'word_3': 'sports',
    'word_4': 'helicopter',
    'word_5': 'witcher',
    'word_6': 'cricket',
    'word_7': 'football',
    'word_8': 'chess',
    'word_9': 'friends',
    'word_10': 'gameofthrones',
}

hangman =(
    """
    x-------x
    """,
    """
    x-------x
    |
    |
    |
    |
    |
    """,
    """
    x-------x
    |       |
    |       0
    |
    |
    |
    """,
    """
    x-------x
    |       |
    |       0
    |       |
    |
    |
    """,
    """
    x-------x
    |       |
    |       0
    |      /|\\
    |
    |
    """,
    """
    x-------x
    |       |
    |       0
    |      /|\\
    |      /
    |
    """,
    """
    x-------x
    |       |
    |       0
    |      /|\\
    |      / \\
    |
    GAME OVER
    """
)

Max = len(hangman) - 1
print(Max)
word_list_num = len(word_list)
print(word_list_num)
num = randint(1, word_list_num)
print(num)
num_string = str(num)
words = 'word_{}'.format(num)
print(words)
word_to_guess = word_list[words]
print("Gussed word: ", word_to_guess)
hidden = ['_'] * len(word_to_guess)
letter_gussed = []

def hangmanGame():
    hangman_size = 0
    print("Hangman")
    word_arr = list(word_to_guess)
    print("array: ", word_arr)
    
    while hangman_size < Max:
        print(str(hidden))
        user_guess = input("Guess a letter...")
        
        if user_guess in letter_gussed:
            print("you already gussed that.Try again...")
            user_guess = user_guess = input("Guess again another letter...")
            
        if user_guess in word_arr:
            print("yea yea you gussed it right..see whether you can finish it")
            
            for num in range(len(word_arr)):
                if user_guess == word_arr[num]:
                    hidden[num] = user_guess
                    if hidden.count('_') == 0:
                        print("you win! finally you did something")
                        quit()
        else:
            print('{}..Really thats the best you can do..see again by trying'.format(user_guess))
            hangman_size += 1
            print(hangman[hangman_size])
            
hangmanGame()
```
