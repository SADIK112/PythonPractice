```python
def plot_corr(data,size=20):
    corr = data.corr()
    fig, ax = plt.subplots(figsize=(size,size))
    ax.matshow(corr)
    cax = ax.matshow(corr, interpolation='nearest')
    fig.colorbar(cax)
    plt.xticks(range(len(corr.columns)),corr.columns,rotation=17)
    plt.yticks(range(len(corr.columns)),corr.columns)
plot_corr(data)
    
```

```python
from sklearn.model_selection import train_test_split

feature_col_names = data.columns[:-1]

predicted_class_name = data.columns[-1:]

X = data[feature_col_names].values

Y = data[predicted_class_name].values

X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.30,random_state=42)

print("train data accuracy: {0}".format(len(X_train)/len(data)))
```

```python
for column in data.columns:
    print("Number if zeroes in {1:12} : {0:>4}".format(len(data.loc[data[column]==0]),column))
```

```python
from sklearn.preprocessing import Imputer

X_train[45]
```

```python
fill_0 = Imputer(missing_values=0, strategy='mean',axis=0)

X_train = fill_0.fit_transform(X_train)
X_test = fill_0.fit_transform(X_test)

X_train[45]
```

```python
from sklearn.naive_bayes import GaussianNB

nb_model = GaussianNB()

nb_model.fit(X_train,Y_train.ravel())
```

```python
nb_predict_train = nb_model.predict(X_train)

from sklearn import metrics

metrics.accuracy_score(Y_train, nb_predict_train)
```

```python
nb_predict_test = nb_model.predict(X_test)

metrics.accuracy_score(Y_test, nb_predict_test)
```

```python
conf_mx = metrics.confusion_matrix(Y_test,nb_predict_test)

print(conf_mx)
```

```python
print(metrics.classification_report(Y_test,nb_predict_test))
```

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train,Y_train.ravel())
```

```python
rf_predict_train = rf_model.predict(X_train)
print(metrics.accuracy_score(Y_train, rf_predict_train))
```

```python
rf_predict_test = rf_model.predict(X_test)
print(metrics.accuracy_score(Y_test, rf_predict_test))
```

```python
print(metrics.classification_report(Y_test,rf_predict_test))
```

```python
from sklearn.linear_model import LogisticRegression

lr_model = LogisticRegression(C=0.7, random_state=42)

lr_model.fit(X_train, Y_train.ravel())

lr_predict_test = lr_model.predict(X_test)

print(metrics.confusion_matrix(Y_test, lr_predict_test))

print(metrics.classification_report(Y_test, lr_predict_test))
```

```python
C_start= 0.1
C_end=5
C_inc=0.1

C_values, recall_scores = [], []

C_val=C_start

best_recall_score = 0

while C_val < C_end:
    C_values.append(C_val)
    lr_model_loop = LogisticRegression(C=C_val,random_state=42)
    lr_model_loop.fit(X_train,Y_train.ravel())
    lr_predict_loop_test = lr_model_loop.predict(X_test)
    recall_score = metrics.recall_score(Y_test,lr_predict_loop_test)
    recall_scores.append(recall_score)
    if ( recall_score > best_recall_score):
        best_recall_score = recall_score
        best_lr_predict_test = lr_predict_loop_test

    C_val=C_val+C_inc




plt.plot(C_values,recall_scores,"-")
```
