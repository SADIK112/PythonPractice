```python
import time

answer_A = ['A' , 'a']
answer_B = ['B' , 'b']
answer_C = ['C' , 'c']

answer_Y = ['Y', 'y' , 'yes', 'Yes']
answer_N = ['N', 'n' , 'no', 'No']

sword = 0
flower = 0

required = ("\nUse only A, B, or C\n")

def intro():
    print("After a drunken night out with friends, you awaken the next morning in a thick, dank forest." 
    "Head spinning and fighting the urge to vomit, you stand and marvel at your new, unfamiliar setting." 
    "The peace quickly fades when you hear a grotesque sound emitting behind you."
    "A slobbering orc is running towards you. You will:")
    time.sleep(1)
    print("""
             A. Grab a nearby rock and throw it at the orc
             B. Lie down and wait to be mauled
             C. Run""")
    choice = input(">>> ")
    if choice in answer_A:
        option_rock()
    elif choice in answer_B:
        print("Welp, that was quick. You died!")
    elif choice in answer_C:
        option_run()
    else:
        print(required)
        intro()
        
        
def option_rock():
    print("The orc is stunned, but regains control. He begins running towards you again. Will you:")
    time.sleep(1)
    print("""
             A. Run
             B. Throw another rock
             C. Run towards a nearby cave""")
    choice = input(">>> ")
    if choice in answer_A:
        option_run()
    elif choice in answer_B:
        time.sleep(1)
        print("You decided to throw another rock, as if the first rock thrown did much damage."
            "\n\n The rock flew well over the orcs head. You missed. You died!")
    elif choice in answer_C:
        time.sleep(1)
        print("You were hesitant, since the cave was dark and ominous." 
              "Before you fully enter, you notice a shiny sword on the ground."
              "\n\nDo you pick up a sword. Y/N?")
        choice = input(">>> ")
        if choice in answer_Y:
            sword = 1
        else:
            sword = 0
        if sword > 0:
            time.sleep(1)
            print("""
                     What do you do next?
                     
                     A. Hide in silence
                     B. Fight
                     C. Run""")
            choice = input(">>> ")
            if choice in answer_A:
                time.sleep(1)
                print("Really? You're going to hide in the dark? I think orcs can see very well in the dark, right?"
                      "\n\nNot sure, but I'm going with YES, so...You died!")
            elif choice in answer_B:
                time.sleep(1)
                print("You laid in wait. The shimmering sword attracted the orc,"
                          "which thought you were no match." 
                          "As he walked closer and closer, your heart beat rapidly."
                          "As the orc reached out to grab the sword, you pierced the blade into its chest."
                          "\n\nYou survived!")
            elif choice in answer_C:
                time.sleep(1)
                option_run()
            
        else:
            time.sleep(1)
            print("\nYou should have picked up that sword. You're "
                "defenseless. \n\nYou died!")
    else:
        print(required)
        intro()
        
def option_run():
    print("You run as quickly as possible, but the orc's speed is too great. You will:")
    time.sleep(1)
    print("""
             A. Hide behind boulder
             B. Trapped, so you fight
             C. Run towards an abandoned town""")
    choice = input(">>> ")
    if choice in answer_A:
        print("You're easily spotted. You died!")
    elif choice in answer_B:
        print("You're no match for an orc. You died!")
    elif choice in answer_C:
        print("While frantically running, you notice a rusted sword lying in the mud."
              "You quickly reach down and grab it, but miss. You try to calm your heavy breathing"
              "as you hide behind a delapitated building, waiting for the orc to come charging around the corner." 
              "You notice a purple flower near your foot. Do you pick it up? Y/N")
        choice = input(">>> ")
        if choice in answer_Y:
            flower = 1
        else: 
            flower = 0
        if flower > 0:
            print("You hear its heavy footsteps and ready yourself for the impending orc.")
            time.sleep(1)
            print("You quickly hold out the purple flower, somehow hoping it will stop the orc"
                  "It does! The orc was looking for love."
                  "\n\n This got weird, but you survived!")
            
        else:
            print("Maybe you should have picked up the flower."
                  "\n\n You died!")
intro()
            
```

```python
from random import randint

print("Hello whats your name... ")
name = input()

guessTaken = 0;

number = randint(1 , 20)

print("Hello " + name + " guess a number between 1 and 20")

while guessTaken < 6:
    print("take a guess... ")
    guess = int(input())
    
    guessTaken += 1
    
    if guess > number:
        print("Your guess is too high.Try again... ")
        
    if guess < number:
        print("Your guess is too low.Try again... ")

    if guess == number:
        break

if guess == number:
    myNumber = str(guessTaken)
    print("Congrats you gussed my number in " + myNumber + " guesses")

if guess != number:
    number = str(number)
    print("sorry! I gussed the number " + number)
```
